**Project Objective:**
Build a movie recommendation system that suggests movies to users based on their preferences or previous ratings.

Data Source: https://raw.githubusercontent.com/YBI-Foundation/Dataset/main/Movies%20Recommendation.csv

```python
#Importing Library
import pandas as pd
import numpy as np
```

```python
#Data Preprocessing
data_features = data[['Movie_Genre','Movie_Keywords','Movie_Cast','Movie_Overview','Movie_Language',]].fillna('')
```

Selected five existing features for recommending movies.

```python
data_features.shape
```

```python
data_features
```

```python
#Define Target Variable (y) and Feature Variables (X)
x = data_features['Movie_Genre']+' '+data_features['Movie_Keywords']+' '+data_features['Movie_Cast']+' '+data_features['Movie_Overview']+' '+data_features['Movie_Language']
```

```python
x
```

```python
x.shape
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfdif = TfidfVectorizer()
```

```python
x = tfdif.fit_transform(x)
```

```python
x.shape
```

```python
print(x)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
Similarity_Score = cosine_similarity(x)
```

```python
Similarity_Score
```

```python
Similarity_Score.shape
```

```python
Favourite_Movie_Name = input("Enter Your favourite movie name: ")
```

```python
All_Movies_Title_List = data['Movie_Title'].tolist()
```

```python
import difflib
```

```python
Movie_Recommendation = difflib.get_close_matches(Favourite_Movie_Name,All_Movies_Title_List)
print(Movie_Recommendation)
```

```python
Close_Match = Movie_Recommendation[0]
print(Close_Match)
```

```python
Index_Of_Close_Match_Movie = data[data.Movie_Title == Close_Match]['Movie_ID'].values[0]
print(Index_Of_Close_Match_Movie)
```

```python
#getting list of similar moves
Recommendation_Score = list(enumerate(Similarity_Score[Index_Of_Close_Match_Movie]))
print(Recommendation_Score)
```

```python
len(Recommendation_Score)
```

```python
#sorting movies based on similarity score
Sorted_Similar_Movies = sorted(Recommendation_Score, key = lambda x:x[1], reverse = True)
print(Sorted_Similar_Movies)
```

```python
#print name of similar movies based on index
print("Top 30 movies suggested for you: \n")

i = 1
for movie in Sorted_Similar_Movies:
  index = movie[0]
  title_from_index = data[data.index == index]['Movie_Title'].values[0]
  if(i<31):
    print(i, '.', title_from_index)
    i+=1
```

```python
#Top 10 Movie Recommendation System
Movie_name = input('Enter your favourite movie name: ')
list_of_all_titles = data['Movie_Title'].tolist()
Find_Close_Match = difflib.get_close_matches(Movie_name, list_of_all_titles)
Close_Match = Find_Close_Match[0]
Index_Of_Close_Match_Movie = data[data.Movie_Title == Close_Match]['Movie_ID'].values[0]
Recommendation_Score = list(enumerate(Similarity_Score[Index_Of_Close_Match_Movie]))
sorted_similar_movies = sorted(Recommendation_Score, key = lambda x:x[1], reverse= True)
print("Top 10 Movies Suggested for You: \n")
i = 1
for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = data[data.Movie_ID == index]['Movie_Title'].values
  if(i<11):
    print(i, '.', title_from_index)
    i+=1
```
